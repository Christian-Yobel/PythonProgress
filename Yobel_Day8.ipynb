8. Displacement adalah jumlah volume total dari seluruh cylinder. Buat kolom baru bernama 'volume_per_cylinder'. 

```python
df['volume_per_cylinder'] = [i for i in df['displacement'] / df['cylinders']]
df
```

```python
df['volume_per_cylinder'] = df['displacement'] / df['cylinders']
```

```python
# Define a function to calculate volume_per_cylinder
def calculate_volume(row):
    return row['displacement'] / row['cylinders']

# Apply the function row-wise
df['volume_per_cylinder'] = df.apply(lambda row: calculate_volume(row), axis=1)
```

```python
df = df.assign(volume_per_cylinder = df['displacement'] / df['cylinders'])
```
