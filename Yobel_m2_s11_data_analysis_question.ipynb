## **No. 6**
Sekarang, dari data mentah di atas, coba buat dataset seperti di bawah ini.  
Perhatikan bahwa:  
* kalian bisa mendapatkan kolom `Gender` dari `Title` yang berasal dari kolom `Name`. Tergantung dari cara kalian membuat kolom gender, jumlah Female/Male yang kalian dapat bisa berbeda 1 dari contoh.
* kolom `Alone` berisi alone/not alone sesuai kondisi penumpang, berdasarkan kolom yang ada.

```python
def group(names):
    gender = re.findall(r', (\w+)\.', names)
    if gender:
        genders = gender[0]
    else:
        genders = 'Unknown'
    title = ["Mr", "Master", "Don", "Rev", "Major", "Sir", "Col", "Capt", "Jonkheer", "Dr"]
    if genders in title:
        return 'Male'
    else:
        return 'Female'

# titanic['Gender'] = titanic['Name'].apply(group)
# titanic

titanic.insert(2, 'Gender', titanic['Name'].apply(group))

# gender_column = titanic['Name'].apply(group)
# titanic.insert(2, 'Gender', gender_column)
titanic
```

```python
def group0(names):
    gender = re.findall(r', (\w+)\.', names)
    if gender:
        return gender[0]
    else:
        return 'Unknown'

titanic.insert(1, 'Title', titanic['Name'].apply(group0))
# titanic['Title'] = titanic['Name'].apply(group0)
# titanic
```

```python
def group2(names):
    last_name = re.findall(r'(\w+), ',names)
    return last_name[0]

titanic.insert(2, 'Last Name', titanic['Name'].apply(group2))
# titanic
# titanic['Last Name'] = titanic['Name'].apply(group2)
titanic
```

```python
def group3(names):
    first_name = re.findall(r'\.\s*(.+?)(?:\(|\,|$)', names)
    if first_name:
        return first_name[0]
    else:
        return "Wrong!"

titanic['Name'] = titanic['Name'].apply(group3)
titanic
```

## **No. 10**

Kemudian, coba lakukan analisis sederhana tentang `survivability` penumpang berdasarkan faktor-faktor seperti gender, umur, kelas sosial dan fare menggunakan functions yang sudah kalian pelajari.

Beri keterangan bagaimana kalian melakukan analisis (kodenya) serta hasil analisis dan kesimpulan kalian dalam bentuk comment (#) atau markdown.

```python
titanic
```

```python
def rate(column):
    pivot1 = pd.pivot_table(data = titanic, index = column, columns = 'Survived', values = 'PassengerId', aggfunc = len)
    pivot1.fillna(0, inplace= True)
    pivot1['Total'] = pivot1.sum(axis = 1)
    pivot1['Survival Rate (%)'] = pivot1[1]/ pivot1['Total'] * 100
    return pivot1
```

```python
rate('Gender')
```

```python
rate('Age')
```

```python
rate('Fare')
```

```python
rate('Pclass')
```
