# **Exercise: Pandas Merging**

Gunakan datasets:
- audi.csv
- hyundi.csv
- toyota.csv

Load ketiga dataset di atas, kemudian simpan masing-masing dataset ke dalam sebuah variable!

```python
import numpy as np
import pandas as pd
```

Dataset: https://www.kaggle.com/adityadesai13/used-car-dataset-ford-and-mercedes?select=toyota.csv

```python
audi = pd.read_csv('audi.csv')
audi.head()
```

```python
hyundai = pd.read_csv('hyundi.csv')
hyundai.head()
```

```python
toyota = pd.read_csv('toyota.csv')
toyota.head()
```

## **No.1**

- Jika Anda perhatikan, nama kolom 'tax(£)' di dataset hyundai berbeda dengan di dataset lainnya.
- Ubah nama kolom tersebut menjadi 'tax' agar sama dengan di dataset yang lain.

```python
hyundai.rename(columns= {'tax(£)': 'tax'}, inplace = True)
hyundai
```

## **No.2**

- Buat kolom baru bernama 'brand' dengan isi nama brand mobilnya, letakkan pada kolom paling kiri. 
- Misalnya untuk data frame audi, isi kolom ‘brand’ dengan ‘Audi’!

```python
# audi.insert(0, 'brand', 'Audi')
audi
```

```python
toyota.insert(0,'brand', 'Toyota')
toyota
```

```python
hyundai.insert(0, 'brand', 'Hyundai')
hyundai
```

## **No.3**

- Gabungkan ketiga data frame tersebut menjadi sebuah data frame bernama df_car. Atur agar indexnya tidak ada yang berulang.


```python
df_car = pd.concat([audi, toyota, hyundai], ignore_index = True)
df_car
```

```python
# # Concatenate the underlying numpy arrays
# data_concat = np.concatenate([df1.values, df2.values])

# # Create a new DataFrame
# df_concat = pd.DataFrame(data_concat, columns=df1.columns)

df_car2 = np.concatenate([audi.values, toyota.values, hyundai.values])
df_car4 = pd.DataFrame(np.concatenate([audi.values, toyota.values, hyundai.values]), columns=audi.columns)
df_car2
```

## **No.4**

- Pada kolom 'model' di sebagian data, ada whitespace tambahan di awal atau di akhir yang tidak berguna. 
- Hapus whitespace tersebut dari semua data pada kolom 'model'!

```python
df_car['model'] = df_car['model'].str.strip()
```

```python
len(df_car['model'][0])
```

```python
df_car.loc[0, 'model']
```

## **No.5**

- Dari df_car, tampilkan rata-rata pajak/tax berdasarkan brand dan transmission. Lalu urutkan dari tax yang paling tinggi.


```python
df_car.groupby(['brand','transmission'])[['tax']].mean().sort_values('tax', ascending=False)
```

## **No.6**

- Anda diminta untuk mencari mobil Toyota Camry yang harganya (price) di bawah rata-rata harga Camry dan jarak tempuhnya (mileage) di bawah rata-rata mileage Camry juga. 
- Tampilkan mobil yang sesuai dengan kriteria tersebut!

```python
camry = df_car[(df_car['model'] == 'Camry')]
camry
```

```python
camry[(camry['mileage'] < camry['mileage'].mean()) & (camry['price'] < camry['price'].mean())][['brand']]
```
