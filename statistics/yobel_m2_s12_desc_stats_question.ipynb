5. Apa yang dimaksud dengan Measure of Spread? Tampilkan Measure of Spread dari kolom 'price'!

```python
# Measure of Spread is a way to explain the data through describing the spread of a group of data
# Std
df_car['price'].std()
df_car['price'].var()
# Range
range = df_car['price'].max() - df_car['price'].min()
range
# Interquartile Range (IQR)
q75, q25 = df_car['price'].quantile([0.75, 0.25])
iqr = q75 - q25
iqr
iqr2 = df_car['price'].describe()['75%'] - df_car['price'].describe()['25%']
iqr2
Q1 = df_car['price'].describe()['25%']
Q3 = df_car['price'].describe()['75%']
```

6. Apakah ada Outlier pada kolom 'price'? Kalau ada, berapa banyak data yang merupakan outliers?

```python
lower_bound = Q1 - 1.5 * iqr
upper_bound = Q3 + 1.5 * iqr
outliers = df_car[(df_car['price'] > upper_bound) | (df_car['price'] < lower_bound)]
outliers
lower_outliers = df_car[(df_car['price'] < lower_bound)]
upper_outliers = df_car[(df_car['price'] > upper_bound)]
upper_outliers
```

7. Variabel/kolom apa saja yang memiliki distribusi skewed ke kanan?

```python
def skews(df, column, output='all'):
    Q1 = df_car[column].describe()['25%']
    Q3 = df_car[column].describe()['75%']
    iqr = Q3 - Q1
    lower_bound = Q1 - 1.5 * iqr
    upper_bound = Q3 + 1.5 * iqr
    outliers = df[(df[column] > upper_bound) | (df[column] < lower_bound)]
    lower_outliers = df[(df[column] < lower_bound)]
    upper_outliers = df[(df[column] > upper_bound)]
    if output == 'all':
        return outliers
    elif output == 'lower':
        return lower_outliers
    elif output == 'upper':
        return upper_outliers

skews(df_car, 'mileage', 'lower')  # 0 data
skews(df_car, 'mileage', 'upper')  # 791 data so it's a right skew 

skews(df_car, 'tax', 'lower')      # 0 data
skews(df_car, 'tax', 'upper')      # 122 Data so it's a right skew

skews(df_car, 'mpg', 'lower')      # 15 data
skews(df_car, 'mpg', 'upper')      # 134 data so it's a right skew

skews(df_car, 'engineSize', 'lower')   # 110 data
skews(df_car, 'engineSize', 'upper')   # 1396 data so it's a right skew
```
